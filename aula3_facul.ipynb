```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Criando dados: 7 dias da semana x 6 períodos do dia
dias = ["Seg", "Ter", "Qua", "Qui", "Sex", "Sáb", "Dom"]
horas = ["Manhã", "Tarde", "Noite", "Madrugada", "Almoço", "Happy Hour"]
vendas = np.random.randint(0, 50, (7, 6))  # Gera valores aleatórios de vendas
# Criando o heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(vendas, annot=True, cmap="coolwarm", xticklabels=horas, yticklabels=dias)
plt.title("Heatmap de Vendas por Dia e Período")
plt.xlabel("Período do Dia")
plt.ylabel("Dia da Semana")
plt.show()


#matriz de correlação
```

```python
# prompt: gerar um grafico de barras, usando o "type 1 " e a quantidade de pokemons do pokedex.csv

import matplotlib.pyplot as plt
# Contagem de Pokémon por tipo
type1_counts = df['Type 1'].value_counts()

# Criar o gráfico de barras
plt.figure(figsize=(12, 6))  # Ajustar o tamanho da figura para melhor visualização
plt.bar(type1_counts.index, type1_counts.values, color='skyblue')
plt.xlabel('Tipo 1', fontsize=12)
plt.ylabel('Quantidade de Pokémons', fontsize=12)
plt.title('Quantidade de Pokémons por Tipo 1', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10) # Rotacionar os rótulos do eixo x para melhor legibilidade
plt.tight_layout() # Ajusta o layout para evitar que os rótulos se sobreponham
plt.show()
```

```python
# prompt: grafico de dispersao entre hp e attack

import matplotlib.pyplot as plt
# Criar o gráfico de dispersão
plt.figure(figsize=(8, 6))
plt.scatter(df['HP'], df['Attack'], alpha=0.7, color='green')  # alpha para transparência dos pontos
plt.xlabel('HP', fontsize=12)
plt.ylabel('Attack', fontsize=12)
plt.title('Gráfico de Dispersão: HP vs. Attack', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5) # Adicionar grade ao gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Ordenar o DataFrame por 'Attack'
df_sorted = df.sort_values(by='Attack', ascending=True)

# Criar o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(df_sorted['Attack'].values, marker='o', linestyle='-', color='b', markersize=6)

# Rótulos e título
plt.xlabel('Índice (após ordenação por Attack)')
plt.ylabel('Attack')
plt.title('Evolução do Ataque (Ordenado)')

# Ajuste fino do layout
plt.tight_layout()
plt.show()
```
